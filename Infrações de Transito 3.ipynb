```python
gravidadeano = infracao.groupby(["ano_cometimento", "grav_tipo"])["grav_tipo"].count()

gravidadeano
```

## No gráfico  podemos observar, que a infração média e muito superior em nível de quantidade que as outras infrações , as outras infrações podemos observar pelo gráfico que tem  quantidades próximas. 

```python
gravidadeano.plot(kind='bar', subplots=True, layout=(4, 4), figsize=(20, 20), sharex=True )
plt.show()
```

```python
maiores_infracao = infracao.groupby(["tipo_infracao","descricao"]).agg({'tipo_infracao':['count']}).reset_index()
maiores_infracao.columns = ['Codigo','Descricao','Quantidade']
maiores_infracao = maiores_infracao.sort_values(by=['Quantidade'],ascending=False)
maiores_infracao = maiores_infracao [0:10]
maiores_infracao
```

## O segundo gráfico podemos observar as 10 multas mais aplicadas nos motorista , e com quase um milhão a mais de multas aplicadas que a segunda está “Transitar em velocidade superior à máxima permitida”, com isso podemos observar que os motoristas não estão respeitando o limite de velocidade das pistas do Distrito Federal

```python
maiores_infracao.plot.bar(x ='Codigo')
plt.show()
```

```python
infracao_leve = infracao[["tipo_infracao","descricao","grav_tipo"]]
infracao_leve = infracao_leve[infracao_leve['grav_tipo']=='Leve']
infracao_leve = infracao_leve.groupby(["tipo_infracao","descricao"]).agg({'tipo_infracao':['count']}).reset_index()
infracao_leve.columns = ['codigo','Descricao','Quantidade']
infracao_leve = infracao_leve.sort_values(by=['Quantidade'],ascending=False)
infracao_leve = infracao_leve[0:4]
infracao_leve
```

## O terceiro gráfico podemos observar que eu selecionei as 4 multas de gravidade leve mais aplicadas, e com mais de 300mil as mais que a segunda colocada está “Transitar com o veículo na faixa ou na pista da direita regulamentada como de circulação exclusiva para determinado tipo de veículo, exceto para acesso a imóveis lindeiros ou conversões a direita.’ com isso observamos que os motoristas não respeitam as faixas exclusivas.
